```python
#imports
import pandas as pd
import time
from pathlib import Path
from datetime import datetime
import sqlite3
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# list of db table names to report
sset_list = ['MID', '10E', 'BRO', 'DGM', 'DMU', 'ELD', 'GTC', 'MAT', 'MOM', 'NEO', 'ONE', 'RTR', 'SNC', 'VOW', 'WOE']
```

```python
# build db conection
price =  Path("../Resources/inventory.db")
con = sqlite3.connect(price)
cur = con.cursor()
totals_price =[]
total_cards = []
```

```python
# work loop
for i in sset_list:
# load table to df    
    db_df = pd.read_sql_query(f"SELECT * FROM '{i}'", con)
    db_df = db_df.fillna("0")
# clean data    
    for i in  range(len(db_df)):
        x = db_df.loc[i,['MktPrice']] 
        x = x['MktPrice']
        x = x.replace("$", "")
        x = x.replace("None", "")
        db_df['MktPrice'][i] = x
    db_df["MktPrice"] = pd.to_numeric(db_df["MktPrice"])
    db_df["quantity"] = pd.to_numeric(db_df["quantity"])
# do math
    db_df['total'] = (db_df["quantity"]) * (db_df["MktPrice"])
    total = db_df['total'].sum()
    card_count = db_df['quantity'].sum()                   # calcuate for card set; total cards, total inventory market value
    totals_price.append(total)
    total_cards.append(card_count)
con.close()
```

```python
report_df =  pd.DataFrame(zip(sset_list, total_cards, totals_price), columns=['set', 'cards', 'price'])
report_df['date']= datetime.today().strftime('%Y-%m-%d')  # add todays date as column
report_df
```

```python
# creat pie chart of top 7 sets in value and collapse all others to other field
pie_df = report_df.sort_values(by='price', ascending=False) # sort values
pie_df2 = pie_df.iloc[0:7]     # slice top seven
other = pie_df.iloc[7:len(pie_df)]  # slice others
otherprice= other["price"].sum()    # calculate others total price
othercards = other["cards"].sum()    # calculate others total cards
pie_df2.loc[-1] = ['other', othercards, otherprice, datetime.today()]     # add row for others
pie_df2.head(8)
```

```python
# graph by showing inventory makeup by set
my_data = pie_df2['price']
my_labels = pie_df2['set'] 
plt.pie(my_data, labels=my_labels, autopct="%1.1f%%", radius =1.5)
plt.savefig('..\Resources\pie.png')
plt.show()
```
