```python
#open predict csv and actual and combine
# open csv to pandas df 
file_path = Path("resources\m1VOW-pre.csv")
m1a_predic = pd.read_csv(file_path)

file_path = Path("resources\m2VOW-pre.csv")
m2a_predic = pd.read_csv(file_path)

file_path = Path("resources\m3VOW-pre.csv")
m3a_predic = pd.read_csv(file_path)


file_path = Path("resources\_vow_actual.csv")
actual = pd.read_csv(file_path)

m1a_predic = m1a_predic.rename(columns={'Price Prediction': 'Machine1'})
m1a_predic['actual price'] = actual["price"]
m1a_predic['Machine2'] = m2a_predic ["Price Prediction"]
m1a_predic['Machine3'] = m3a_predic ["Price Prediction"]
m1a_predic.head(3)
```

```python
#predict accuracy for the three machines for predicted values for set VOW
accuracy_df = m1a_predic[['week_num','Machine1', 'Machine2', 'Machine3', 'actual price']].copy()
machine_list = ['Machine1', 'Machine2', 'Machine3']

for x in machine_list:
    accuracy_df[f'{x}per'] = (accuracy_df['actual price'] / accuracy_df[x] * 100 -100).round(2)
for x in machine_list:
    nonegs = []
    for i in accuracy_df[f'{x}per']:
        if i >= 0:
            nonegs.append(i)
        if i < 0:
            i = i * -1
            nonegs.append(i)
    accuracy_df[f'{x}n-'] = nonegs

accuracy_df.head(25)
```

```python
# calculate average error
for x in machine_list:
    avg=  accuracy_df[f'{x}n-'].mean()
    print(f'the {x} average error is {avg}')
```

```python
# make and save graph for gard set with actual price and 3 machines 
plt.plot(m1a_predic['week_num'], m1a_predic['Machine1'], color='blue', label='Machine 1')
plt.plot(m1a_predic['week_num'], m1a_predic['actual price'], color='red', label='actual')
plt.plot(m1a_predic['week_num'], m1a_predic['Machine2'], color='green', label='Machine 2')
plt.plot(m1a_predic['week_num'], m1a_predic['Machine3'], color='yellow', label='Machine 3')
plt.xlabel('Week Number')
plt.ylabel('Price')
plt.legend(loc="upper right")
plt.title('VOW Data')
plt.savefig('resources\Images\VOW.png')
plt.show
```

```python
# plot a grouped bar graph to show error
species = ("Machine1", "Machine2", "Machine3")
error = {
    'Training error': (4.64, 4.35, 4.04),
    'Prediction error': (10.67, 8.82, 10.07),
    
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in error.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percent')
ax.set_title('Percent Error by data set')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 12)
plt.savefig('resources\Images\_bar.png')
plt.show()
```
