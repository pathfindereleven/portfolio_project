```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
import time
import requests
import os
import csv
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from pathlib import Path
from datetime import datetime
```

```python
from splinter import Browser
from selenium.webdriver.chrome.service import Service

my_service = Service(executable_path=r'C:\Users\pc\OneDrive\Desktop\TCG-scraper\chromedriver.exe')
browser = Browser('chrome', service=my_service)
```

```python

sset = "dragons_maze_csv"
```

```python
# visit browser and make soup
url ="https://shop.tcgplayer.com/price-guide/magic/dragons-maze"
browser.visit(url)
```

```python
req = browser.html
time.sleep(3)
soup = BeautifulSoup(req, "html")
```

```python
print(soup)
```

```python
names= []
title_tags = soup.find_all('a', class_ = "pdp-url")
#title_tags
for i in title_tags:
   names.append(i.text)
```

```python
prices =[]
price = soup.find_all('td', class_ = "tcg-table-body__cell tcg-table-body__cell--align-right")
#price
count = -1
for i in price:
    count = count +1
    if  count%2 > 0 :
        prices.append(i.text)
```

```python
# turn lists into new data frame
price_list= zip(names, prices)
mkt_df = pd.DataFrame(price_list) 
mkt_df
mkt_df.columns = ['Name', 'Market Price']
mkt_df.head(3)
```

```python
mkt_df['date']= datetime.today().strftime('%Y-%m-%d')
mkt_df.head(3)
```

```python
# Read the sales data
file_path = Path("dragons_maze_csv.csv")
set_df = pd.read_csv(file_path)

set_df["MktPrice"] = 0
set_df.head(1)
```

```python
# Match inventory df  to scrapped data and append current Market price to inventory df
count = 0
while count < len(set_df): # next itteration if loop fails 
    try:
        for i in range(count , len(set_df)):
            search = set_df.loc[i,['card_name']]   # 
            search = search['card_name']
            search_df = mkt_df.loc[mkt_df['Name'] == search]
                           # break if no match
            search_df.reset_index(drop=True, inplace=True)
            search_result = search_df["Market Price"][0]
            set_df.loc[i,['MktPrice']] = search_result
            count = count + 1
    except: # if no match 
            set_df.loc[i,['MktPrice']] = "na"
            count = count + 1
    
#
```
