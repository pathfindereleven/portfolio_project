```python
import sqlite3
from pathlib import Path
import pandas as pd
```

```python
price =  Path("price.db")
con = sqlite3.connect(price)
```

```python
cur = con.cursor()
```

```python
q = cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
for name in q.fetchall():
    print(name[0])
```
